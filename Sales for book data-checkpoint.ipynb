```python
import sqlite3
import random

conn = sqlite3.connect(r"C:\Users\jakes\Downloads\books.db")
cursor = conn.cursor()

query = """
SELECT 
    books_clean.title,
	books_clean.average_rating,
    SUM(sales.quantity) AS Total_sales_per_book
FROM sales
JOIN books_clean 
ON sales.book_id = books_clean.book_id
Group by books_clean.title
Order by Total_sales_per_book DESC ;
"""
cursor.execute(query)
results = cursor.fetchall()

conn.close()

```

```python
average_rating=[column[1]for column in results]
total_sales=[column[2] for column in results]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
#plt.xlim(4.5,5)
plt.scatter(average_rating,total_sales)

plt.xlabel("Average rating")
plt.ylabel("Total sales")
plt.title("Total Sales against average ratings")

#plt.xticks(rotation=45)
#plt.tight_layout()

plt.show()
```

```python
import sqlite3
import random

conn = sqlite3.connect(r"C:\Users\jakes\Downloads\books.db")
cursor = conn.cursor()

query = """
SELECT 
    sales.sale_date,
    SUM(sales.quantity) AS Total_sales_per_day
FROM sales
Group by sales.sale_date
Order by sales.sale_date ASC;
"""
cursor.execute(query)
results = cursor.fetchall()

conn.close()
```

```python
sale_date=[column[0]for column in results]
total_sales=[column[1] for column in results]
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from datetime import datetime
plt.figure(figsize=(10, 6))
plt.bar(sale_date, total_sales)

plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

plt.xlabel("Sale date")
plt.ylabel("Total sales")
plt.title("Total sales in 2024")

#plt.xticks(rotation=45)
#plt.tight_layout()

plt.show()
```
